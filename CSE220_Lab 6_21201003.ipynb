```python
#1(a)
def fac(n):
  if n == 0:
    return 1
  elif n>0:
    return n*fac(n-1)
  else:
    return None
```

```python
#1(b)
def fib(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)
```

```python
#1(c)
def printArrRecursive(arr,idx):
  if idx>=0 and idx<len(arr):
    print(arr[idx],end=" ")
    printArrRecursive(arr,idx+1)
    

arr = [1, 2, 3, 4, 5]
printArrRecursive(arr, 0)

```

```python
#1(d)
def powerN(N,P):
  if P == 0:
    return 1
  else:
    return N*powerN(N,P-1)

print(powerN(3,3))
```

```python
#2(a)
def d2b(n):
    if n == 0:
        return "0"
    elif n == 1:
        return "1"
    else:
        q = n // 2
        r = n % 2
        return d2b(q) + str(r)
print(d2b(2))
```

```python
#2(B)
class Node:
  def __init__(self,e,n=None):
    self.elem = e
    self.next = n
def addList(head):
    if head == None:
      return 0
    else:
      return head.elem + addList(head.next)


head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

sum = addList(head)
print(sum)
```

```python
#3
def hocBuilder(height):
  if height <= 0:
    return 0
  elif height == 1:
    return 8
  else:
    return 5*(height - 1) + hocBuilder(1)

print(hocBuilder(5))

```

```python
#4(a)
def print_pattern(n):
    if n == 1:
        print(n)
        return
    else:
      print_pattern(n-1)
      for i in range(1, n+1):
        print(i, end=' ')
      print()

print_pattern(5)
```

```python
#4(b)
def print_pattern(n, k=1):
    if n == 0:  # base case
        return
    print_pattern(n-1, k+1)  #
    print("  "*(k-1), end='')  
    for i in range(1, n+1):
        print(i, end=' ')
    print()  

print_pattern(5)
```

```python
#5
class FinalQ:
    def print(self, array, idx):
        if idx < len(array):
            profit = self.calcProfit(array[idx], 0)
            print("Investment: ${}  |  Profit: ${}".format(array[idx], profit))
            self.print(array, idx+1)

    def calcProfit(self, investment, profit):
        if investment <= 100000:
            if investment <= 25000:
                return profit
            else:
                return self.calcProfit(investment-100, profit+4.5)
        else:
            if investment <= 125000:
                return self.calcProfit(100000, profit+4.5)
            else:
                return self.calcProfit(125000, profit+4.5) + self.calcProfit(investment-125000, profit+8)


array = [25000, 100000, 250000, 350000]
f = FinalQ()
f.print(array, 0)
```

```python
#PDF Task
class Node:
    def __init__(self, next_node=None, bottom_node=None, val=None):
        self.next = next_node
        self.bottom = bottom_node
        self.val = val

def flattenList(head, output_list):
    if head == None:
        return output_list
    output_list += [head.val]
    if head.bottom != None:
        flattenList(head.bottom, output_list)
    return flattenList(head.next, output_list)
```
