```python
def find_mean(source):
  sum = 0
  for i in range(0,len(source)):
    sum=sum+source[i]
  mean= sum/len(source)
  return mean

def find_standarddeviation(source):
  sum=0
  mean= find_mean(source)
  for j in range(0,len(source)):
      sum = sum+ (source[j]-mean)**2
  sd = (sum/len(source)-1)**0.5
  return sd




def sd_away(n):
  arr= [0,0]
  mean= find_mean(source)
  sd= find_standarddeviation(source)
  sd_above = mean + (n)*sd
  arr[0] = sd_above
  sd_below= mean - (n)*sd
  arr[1] = sd_below
  return arr

source = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]
returned_value = find_mean(source)
sd= find_standarddeviation(source)
d= sd_away(1.5)
print(returned_value)
print(sd)
print(d)
```
