```python
st1= input('Enter 1st string: ')
st2= input('Enter 2nd string: ')
if len(st1) > 10 or len(st2) > 10:
  print('Invalid Input Size')
else:
  a= [['']*10, ['']*10]
  for i in range(len(st1)):
    a[0][i] = st1[i]
  for j in range(len(st2)):
    a[1][j] = st2[j]
  for k in range(len(a)):
    for l in range(len(a[k])):

      if a[k][l] != ''  and ord(a[k][l])>= 65 and ord(a[k][l])<=90:
        uppercase= a[k][l]
        starting_index= l
        print('Uppercase Letter: ', uppercase)
        print('Starting index:', starting_index)

while True:
  user_input= input('Enter command,[Q/q to stop] ')
  if user_input != 'q' and user_input != 'Q':
    temp1= a[0][0]
    temp2= a[1][len(a[1])-1]
    for index in range(1,len(a[0])):
      a[0][index-1] = a[0][index]
    a[0][len(a[0])-1]= temp1
    for idx in range(len(a[1])-1,0,-1):
      a[1][idx] = a[1][idx-1]
    a[1][0] = temp2
  else:
    break
  for x in range(len(a[0])):
    print(a[0][x], end='')
  print()
    
  for y in range(len(a[1])):
   print(a[1][y],end='')
  print()

```
