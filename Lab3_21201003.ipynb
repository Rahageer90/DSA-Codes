```python
# Run This cell Second
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n
```

```python
# To you coding in this cell
class LinkedList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array. head will refer
  #  to the Node that contains the element from a[0]
  #  Else Sets the value of head. head will refer
  #  to the given LinkedList

  # Hint: Use the type() function to determine the data type of a
    self.head = None
    tail = None
    if type(a) == list:
      self.size = len(a)
      for i in range(0,len(a)):
        n = Node(a[i], None)
        if self.head == None :
          self.head = n
          tail = n
        else:
          tail.next = n
          tail = n
    else:
      self.head = a
    # To Do
    
  
  # Traverse elements in the list.
  # This method is done for you. Do not change this method.
  def traverseList(self):
    s = ''
    temp = self.head
    while temp != None:
      if temp.next != None:
        s += str(temp.element) + " "
      else:
        s += str(temp.element)
      temp = temp.next
    return s
  
  # Count the number of nodes in the list and return the total number
  def countNode(self):
    # To Do
    count = 0
    temp = self.head
    while temp != None:
      count+=1
      temp = temp.next
    return count
  
  # returns the reference of the Node at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    temp = self.head
    c = 0
    obj = None
    while temp != None:
      if c == idx:
        obj = temp
        break
      temp = temp.next
      c+=1
    return obj
  
  # returns the element of the Node at the given index. For invalid idx return None.
  def get(self, idx):
    # To Do
    c = 0
    temp = self.head
    obj = None
    while temp != None:
      if c == idx:
        obj = temp.element
        break
      temp = temp.next
      c+=1
    if obj == None:
      print('None')
    return obj
  
  # updates the element of the Node at the given index. 
  # Returns the old element that was replaced. For invalid index return None.
  # parameter: index, element
  def set(self, idx, elem):
    # To Do
    c = 0
    temp = self.head
    while temp != None:
      if c == idx:
        obj= temp.element
        temp.element = elem
        return obj
      c+=1
      temp = temp.next

  # returns the index of the Node containing the given element.
  # if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    temp = self.head
    c= 0
    while temp != None:
      if elem == temp.element:
        return c
      c+=1
      temp = temp.next
    return -1
  
  # returns true if the element exists in the List, return false otherwise.
  def contains(self, elem):
    # To Do
    temp = self.head
    while temp!= None:
      if temp.element == elem:
        return True
      temp = temp.next
    return False

  # Makes a duplicate copy of the given List. Returns the reference of the duplicate list.
  def copyList(self):
    # To Do
    temp = self.head
    n = Node(temp.element, None)
    n2 = n
    while temp != None:
      if temp.next != None:
        x = Node(temp.next.element, None)
        n2.next = x
        n2 = n2.next
      temp = temp.next
    return n

  # Makes a reversed copy of the given List. Returns the head reference of the reversed list.
  def reverseList(self):
    # To Do
    newhead = None
    temp = self.head
    while temp != None:
      n = temp.next
      temp.next = newhead
      newhead = temp
      temp = n
    return newhead
  
  # inserts Node containing the given element at the given index
  # Check validity of index. If invalid then print "Invalid Index"
  def insert(self, elem, idx):
    # To Do
    newnode = Node(elem, None)
    if idx == 0:
      newnode.next = self.head
      self.head = newnode
    else:
      x = self.nodeAt(idx-1)
      newnode.next = x.next
      x.next = newnode

  # removes Node at the given index. returns element of the removed node.
  # Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    if idx < 0 or idx> self.size:
      return None
    if idx == 0:
      rmv = self.head.element
      self.head = self.head.next
    else:
      x = self.nodeAt(idx-1)
      rmv = x.next.element
      x.next = (x.next).next
    return rmv
    self.size -=1
  
  # Rotates the list to the left by 1 position.
  def rotateLeft(self):
    # To Do
    nh = self.head.next
    temp = nh
    while temp.next != None:
      temp = temp.next
    temp.next = self.head
    self.head.next = None
    self.head = nh
    return self.head
  
  
  # Rotates the list to the right by 1 position.
  def rotateRight(self):
    # To Do
    i = self.head.next
    i2 = self.head
    while i.next != None:
      i = i.next
      i2 = i2.next
    i.next = self.head
    i2.next = None
    self.head = i
    return self.head
```
