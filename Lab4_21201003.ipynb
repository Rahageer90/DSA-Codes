```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    # To Do
    self.head = Node(a[0], None, None)
    self.head.next = self.head
    self.head.prev = self.head
    tail = self.head
    for i in range(1,len(a)):
      n= Node(a[i],None,None)
      tail.next= n
      n.prev= tail
      tail= n
      tail.next= self.head
      self.head.prev= tail
    


  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    # To Do
    temp = self.head
    count = 0
    while temp != None :
      count += 1
      temp = temp.next
      if temp == self.head:
        break
    return count
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    temp = self.head
    while temp != None:
      print(temp.element,end=" ")
      temp = temp.next
      if temp == self.head:
        break
    print()

  # prints the elements in the list backward
  def backwardprint(self):
    temp = self.head.prev
    while temp!= self.head:
      print(temp.element,end=" ")
      temp = temp.prev
    print(self.head.element, end=" ")
    print()
  
  

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    temp= self.head
    count=0
    while temp!= self.head.prev:
      if count== idx:
        return temp
      count+=1
      temp= temp.next
    return None
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    temp= self.head.prev
    count=3
    while temp!= self.head:
      if temp.element== elem:
        return count
      count-=1
      temp= temp.prev
    return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    temp = self.head
    if idx<0 and idx>self.countNode():
      return None
    elif idx == 0 :
      n = Node(elem, self.head, self.head.prev)
      self.head.prev.next = n
      self.head.prev = n
      self.head = n
      return self.head
    else:
      for i in range(0,idx-1):
        temp = temp.next
      n = Node(elem,temp.next,temp)
      temp.next.prev = n
      temp.next= n
      return self.head
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    temp = self.head
    r = None
    if idx<0 and idx> self.countNode():
      return r
    elif idx == 0:
      r = self.head.element
      self.head.next.prev = self.head.prev
      self.head.prev.next = self.head.next
      self.head = self.head.next
      return str(r)
    else:
      for i in range(0,idx-1):
        temp = temp.next
      r = temp.next.element
      temp.next = temp.next.next
      temp.next.prev = temp
      return str(r)

```
