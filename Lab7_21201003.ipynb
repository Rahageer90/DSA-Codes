```python
class KeyIndex:
    def __init__(self, a):
        self.a = a
        min_val = min(a)
        max_val = max(a)

        self.k = [0] * (max_val - min_val + 1) #Finding out range 
        

        for num in a:

            self.k[num - min_val] += 1 #Creating array k
        
            
    def search(self, val):
      if val < min(self.a) or val > max(self.a):
        return False
      if self.k[val - min(self.a)] > 0:
        return True
      return False

    def sort(self):
        for i in range(len(self.a)):
            for j in range(len(self.a)-1):
                if self.a[j] > self.a[j+1]:
                    self.a[j], self.a[j+1] = self.a[j+1], self.a[j]
        return self.a

a = [4, 1, 3, 2, 5, 1, 6, 2]

ki = KeyIndex(a)

print(ki.search(-1))
print(ki.search(0)) 

sort = ki.sort()

print(sort) 
```

```python
def hash_function(s):
    consonants = 'BCDFGHJKLMNPQRSTVWXYZ' 
    consonant_count = 0
    digit_sum = 0
    for i in s.upper():
        if i in consonants:
            consonant_count += 1
    for j in s:
        if j.isdigit():
            digit_sum += int(j)
    return (consonant_count * 24 + digit_sum) % 9
def insert(s,arr=[None]*9):
  idx = hash_function(s)
  
  while arr[idx] != None:
    idx = idx + 1 % len(arr)
  arr[idx] = s


arr = [None] * 9

insert('hello234', arr)
insert('world678', arr)
insert('python135', arr)
insert('programming6676', arr)
insert('hash444', arr)

print(arr)
```
